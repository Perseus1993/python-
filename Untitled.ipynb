```python
# matplotlib基础
from matplotlib import pyplot as plt 


#例1 假设一天中每隔两个小时的气温分别是[15,13,14,5,17,20,25,26,26,27,22,18,15]
```

```python
# x轴
x = range(2, 26, 2)
# y轴
y = [15,13,14.5,17,20,25,26,26,27,22,18,15]

# 尺寸和分辨率
fig = plt.figure(figsize = (15,10), dpi = 80)
plt.plot(x,y)

#设置x轴刻度
plt.xticks(x)

#设置x轴刻度
plt.yticks(range(min(y),max(y)))

plt.show()
plt.savefig("./1.png")
#如果像保存成矢量图，那就把png换成svg格式
```

```python
#例2 假设有10点到12点的每一分钟的气温，绘制一下

# a = [random.randint(20, 35) for i in range(120)]
import random

x = range(0,120)

y = [random.randint(20, 35) for i in range(120)]

fig = plt.figure(figsize = (15, 10), dpi = 80)

#调整x轴

_x_labels = ["10:{}".format(i) for i in range(60)]
_x_labels += ["11:{}".format(i) for i in range(60)]

#为了防止过于密集，取步长为3，并且竖着写
plt.xticks(list(x)[::3],_x_labels[::3], rotation = 90)

#添加描述信息
plt.xlabel("time")
plt.ylabel("temp")
plt.plot(x,y)
plt.show()
```
